```python
import os
import numpy as np
import pandas as pd
import nltk
import copy
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error as mse
import networkx as nx
```

```python
def import_file(file_path):
    # read the contents of the file
    with open(file_path) as f:
        contents = f.readlines()
    
    # separate the first three lines from the rest of the content
    header_info = contents[:3]
    content = contents[3:]
    
    # tokenize the lines of the content
    for i, line in enumerate(content):
        content[i] = nltk.RegexpTokenizer(r"\w+").tokenize(line)
    
    # tokenize the last line of the header_info
    header_info[-1] = nltk.RegexpTokenizer(r"\w+").tokenize(header_info[-1])
    
    # create a pandas dataframe from the content
    df = pd.DataFrame(content[1:], columns=content[0])
    
    # convert the FromNodeId and ToNodeId columns to integers
    df["FromNodeId"] = df["FromNodeId"].map(int)
    df["ToNodeId"] = df["ToNodeId"].map(int)
    
    return header_info, df
```

```python
def create_graph(df):
    # create a directed graph from the pandas dataframe
    G = nx.from_pandas_edgelist(df, source="FromNodeId", target="ToNodeId")
    
    # convert the graph to a directed graph and return it
    return G.to_directed()
```

```python
def page_rank(G, d=0.85, max_iters=100):
    n = G.number_of_nodes()
    rank = {}
    node_val = {}
    error = []
    
    # initialize the rank and node_val dictionaries
    for node in G.nodes():
        rank[node] = 1/n
        node_val[node] = 0
    
    # run the PageRank algorithm for the specified number of iterations
    for _ in range(max_iters):
        # calculate the contribution of each incoming edge to each node's value
        for node in G.nodes():
            out = list(G.out_edges(node))
            n_edges = len(out)
            for edge in out:
                node_val[edge[1]] += rank[edge[0]] / n_edges
        
        # update the rank values for each node
        for node in G.nodes():
            node_val[node] = (1 - d) / n + d * node_val[node]
        
        # calculate the mean squared error between the old and new rank values
        e = mse(list(rank.values()), list(node_val.values()))
        error.append(e)
        
        # update the rank dictionary and reset the node_val dictionary
        for node in G.nodes():
            rank[node] = node_val[node]
            node_val[node] = 0
        
        # stop iterating if the mean squared error is below a certain threshold
        if e < 1e-15:
            break
        
    return error, rank
```

```python
import numpy as np

def updation(G, a, b, norm, type_):

    for node in G.nodes():
        if type_ == "authority":
            edges = G.in_edges(node)
        else:
            edges = G.out_edges(node)
        n_edges = len(edges)
        a[node] = 0
        for edge in edges:
            if type_ == "authority":
                a[node] += b[edge[0]]
            else:
                a[node] += b[edge[1]]
        norm += a[node] ** 2
    norm = np.sqrt(norm)
    for node in G.nodes():
        a[node] /= norm


def HITS(G, max_iters=100):

    n = G.number_of_nodes()
    hub = {}
    authority = {}
    e1 = []
    e2 = []
    
    # initialize the hub and authority scores to 1 for each node
    for node in G.nodes():
        hub[node] = 1
        authority[node] = 1
    
    cur_auth = authority.copy()
    cur_hub = hub.copy()
    
    # iterate until convergence or the maximum number of iterations is reached
    for _ in range(max_iters):
        updation(G, authority, hub, 0, "authority")
        updation(G, hub, authority, 0, "hub")
        
        e1.append(mse(list(authority.values()), list(cur_auth.values())))
        e2.append(mse(list(hub.values()), list(cur_hub.values())))

        cur_auth = authority.copy()
        cur_hub = hub.copy()
        
        if e1[-1] < 1e-20 and e2[-1] < 1e-20:
            break
            
    return authority, hub, e1[1:], e2[1:]
```

```python
def sortDict(d, title):
    d = {key : value for key, value in sorted(d.items(), key=lambda item: item[1], reverse = True)}
    i = 0
    print("top 10 " + title + " scores :")
    for key, val in d.items():
        if i<10:
            print(key,"\t",val)
        i+=1
    print()
    return d
```
